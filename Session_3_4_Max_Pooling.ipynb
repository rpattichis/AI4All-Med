# Max Pooling for Reducing Dimensionality

# Goals

1. Explore Max pooling.
2. Modify Max pooling inputs and parameters.
3. Recognize Max pooling outputs.

# Import libraries

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import seaborn as sns
```

# Set up the input tensor

What makes the input of size (1, 1, 4, 4)?

After running the example, modify the input feature map and explore different outputs.

```python
# Create a single-channel 4x4 feature map (1, 1, 4, 4)
input_tensor = torch.tensor([[[[1.0, 2.0, 3.0, 0.0],
                               [4.0, 5.0, 6.0, 1.0],
                               [7.0, 8.0, 9.0, 2.0],
                               [0.0, 1.0, 2.0, 3.0]]]])

print("Input shape:", input_tensor.shape)
```

# Define the max pooling parameters

1. What do the function inputs mean?
2. What does the stride do?
3. What does the kernel size do?

After running the example below, modify the kernel size and the stide and discuss the relationship between inputs and outputs.

```python
pooled_tensor = F.max_pool2d(input_tensor, kernel_size=2, stride=2)
print("Pooled Output:\n", pooled_tensor)
```

# Visualize the inputs and outputs

```python
def show_tensor(tensor, title=""):
    tensor = tensor.squeeze().numpy()
    plt.figure(figsize=(3, 3))
    sns.heatmap(tensor, annot=True, cbar=False, square=True, fmt=".0f", cmap="YlGnBu")
    plt.title(title)
    plt.axis('off')
    plt.show()

show_tensor(input_tensor, "Original 4×4 Input")
show_tensor(pooled_tensor, "2×2 Max Pooling Output (2×2 blocks)")
```
